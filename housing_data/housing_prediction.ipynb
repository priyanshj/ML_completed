```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

df_test = pd.read_csv("test.csv")
df_test["source"] = "test"

df_train = pd.read_csv("train.csv")
df_train["source"]  = "train"

df = pd.concat([df_test,df_train])

#df = pd.read_csv("sample_submission.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["BsmtFinSF1"].fillna(0,inplace=True)
df["BsmtFinSF2"].fillna(0,inplace=True)
df["BsmtUnfSF"].fillna(df["BsmtUnfSF"].median(),inplace=True)
df["Alley"].fillna("None",inplace=True)
df["PoolQC"].fillna("NA",inplace=True)
df["Fence"].fillna("NA",inplace=True)
df["MiscFeature"].fillna("NA",inplace=True)
df["SaleType"].fillna("WD",inplace=True)
```

```python
Lot_table = df.pivot_table(values="LotFrontage",index="LotArea")
miss_val = df["LotFrontage"].isnull()
df.loc[miss_val,"LotFrontage"] = df.loc[miss_val,"LotArea"].apply(lambda x: Lot_table[x])
```

```python
df["Age_House"] = df["YrSold"] - df["YearRemodAdd"]
location = df.loc(df["GarageCars"].isnull())
```

```python
df["LotFrontage"].fillna(df["LotFrontage"].mean(),inplace=True)
```

```python
df["GarageCars"].value_counts()
df["GarageCars"].fillna(0.0,inplace=True)
df["GarageArea"].value_counts()
df["GarageArea"].fillna(0.0,inplace=True)
df["Garage"] = df["GarageCars"]*df["GarageArea"]
```

```python
df["Floor_Area"] = df["1stFlrSF"] + df["2ndFlrSF"] + df["3SsnPorch"]
df["BsmtFinSF"] = df["BsmtFinSF1"] + df["BsmtFinSF2"] - df["BsmtUnfSF"]
df["BsmtBath"] = df["BsmtFullBath"] + (1/2)*df["BsmtHalfBath"]
```

```python
df["BsmtBath"].value_counts()
df["BsmtBath"].fillna(0.0,inplace=True)
df["TotalBath"] = df["BsmtBath"] + df["FullBath"] + (1/2)*df["HalfBath"]
```

```python
df["Maintainance"] = df["OverallCond"]*df["OverallQual"]
df.head()
```

```python

test_data = df.loc[df["source"] == "test"]
train_data = df.loc[df["source"] == "train"]

mean = train_data["SalePrice"].mean()

sol_mean = test_data[["Id","SalePrice"]]
sol_mean["SalePrice"] = mean

sol_mean.to_csv("solution_base.csv",index=False)
```

```python
target = 'SalePrice'
IDcol = ['Id']

from sklearn import cross_validation,metrics
def modelfit(alg, dtrain, dtest, predictors, target, IDcol, filename):
    #Fit the algorithm on the data
    alg.fit(dtrain[predictors], dtrain[target])
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain[predictors])

    #Perform cross-validation:
    cv_score = cross_validation.cross_val_score(alg, dtrain[predictors], dtrain[target], cv=20, scoring='mean_squared_error')
    cv_score = np.sqrt(np.abs(cv_score))
    
    #Print model report:
    print("\nModel Report")
    print("RMSE : %.4g" % np.sqrt(metrics.mean_squared_error(dtrain[target].values, dtrain_predictions)))
    print("CV Score : Mean - %.4g | Std - %.4g | Min - %.4g | Max - %.4g" % (np.mean(cv_score),np.std(cv_score),np.min(cv_score),np.max(cv_score)))
    
    #Predict on testing data:
    dtest[target] = alg.predict(dtest[predictors])
    
    #Export submission file:
    IDcol.append(target)
    submission = pd.DataFrame({ x: dtest[x] for x in IDcol})
    submission.to_csv(filename, index=False)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso

predictors = ["1stFlrSF","2ndFlrSF","GrLivArea","Floor_Area","TotRmsAbvGrd","Age_House","Maintainance","Garage","BsmtFinSF","LotArea","LotFrontage","MSSubClass","TotalBath","WoodDeckSF"]

alg1 = LinearRegression(normalize=True)
modelfit(alg1,train_data,test_data,predictors,target,IDcol,'alg1.csv')
```
