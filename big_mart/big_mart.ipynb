```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

df_train = pd.read_csv("/home/priyansh/Desktop/shell/big_mart/train.csv")
df_test = pd.read_csv("/home/priyansh/Desktop/shell/big_mart/test.csv")

df_train['source']='train'
df_test['source']='test'

df = pd.concat([df_train,df_test],ignore_index=True)
```

```python
sale = df["Item_Fat_Content"].value_counts()
df["Item_Fat_Content"]=df["Item_Fat_Content"].replace({'LF':'Low Fat','reg':'Regular','low fat':'Low Fat'})
print(df["Item_Fat_Content"].value_counts())
```

```python
df.apply(lambda x: len(x.unique()))
```

```python
weight_table = df.pivot_table(values="Item_Weight",index="Item_Identifier")
miss_val = df["Item_Weight"].isnull()
df.loc[miss_val,'Item_Weight'] = df.loc[miss_val,'Item_Identifier'].apply(lambda x: weight_table[x])
```

```python
df.describe()
```

```python
df.Outlet_Type.unique()
```

```python
df["Outlet_Size"].value_counts()
missing_values = df['Outlet_Size'].isnull() 
df.loc[missing_values,"Outlet_Size"] = df.loc[missing_values,"Outlet_Type"].apply(lambda x: "Medium")
missing_values2 = df['Outlet_Size'].isnull()
df.describe()
```

```python
df.head()
```

```python
visibility = df.pivot_table(values="Item_Visibility",index="Item_Identifier")
missing_val = (df["Item_Visibility"] == 0)
df.loc[missing_val,"Item_Visibility"] = df.loc[missing_val,"Item_Identifier"].apply(lambda x: visibility[x])
```

```python
df.describe()
```

```python
df["Outlet_Years"] = 2013 - df["Outlet_Establishment_Year"]
df["Outlet_Years"].describe()
```

```python
#Get the first two characters of ID:
df['Item_Broad_Category'] = df['Item_Identifier'].apply(lambda x: x[0:2])
df['Item_Broad_Category'] = df['Item_Broad_Category'].map({'FD':'Food','NC':'Non-Consumable','DR':'Drinks'})
df['Item_Broad_Category'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df["Outlet"]=le.fit_transform(df["Outlet_Identifier"])
var_mod = ["Item_Fat_Content","Outlet_Location_Type","Outlet_Size","Item_Broad_Category","Outlet_Type","Outlet"]

for i in var_mod:
    df[i]=le.fit_transform(df[i])
```

```python
pd.get_dummies(df,columns=["Item_Fat_Content","Outlet_Location_Type","Outlet_Size","Item_Broad_Category","Outlet_Type","Outlet"])
```

```python
df.dtypes
```

```python
df.drop(["Outlet_Establishment_Year","Item_Type"],axis=1,inplace=True)

train_data = df.loc[df['source']=="train"]
test_data = df.loc[df['source']=="test"]

#Drop unnecessary columns:
test_data.drop(['Item_Outlet_Sales','source'],axis=1,inplace=True)
train_data.drop(['source'],axis=1,inplace=True)

#Export files as modified versions:
train_data.to_csv("train_modified.csv",index=False)
test_data.to_csv("test_modified.csv",index=False)
```

```python
#baseline solution
mean = train_data['Item_Outlet_Sales'].mean()

sol_1 = test_data[['Item_Identifier','Outlet_Identifier']]
sol_1['Item_Outlet_Sales'] = mean

sol_1.to_csv("solution_base.csv",index=False)
```
